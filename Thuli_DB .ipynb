```python
# Function to categorize body shape
def categorize_body_shape(measurements):
    bust, waist, hips = map(float, measurements.split('-'))

    if abs(bust - hips) < 2 and (bust - waist) >= 7:
        return 'Hourglass'
    elif hips > bust:
        return 'Pear'
    elif bust > hips:
        return 'Apple'
    else:
        return 'Rectangle'

# Applying the function to the dataframe
df['body_shape'] = df['measurements'].apply(categorize_body_shape)
df['name '] = df['name '].str.lower()


# Show the dataframe with body shape appended
print(df)
```

```python
one_hot_body_shape = pd.get_dummies(df['body_shape'], prefix='body_shape').astype(int)

# Combine the one-hot encoded body shapes with names
body_shape_encoded = pd.concat([df['name '],one_hot_body_shape], axis=1)
```

```python
# Sample query vector
#query_vector = np.array([0.06255876, 0.18371858, 0.02128122, 0.41576535, 0.32767248, 0.13637275])
# Function to find the closest point in Qdrant
def find_closest_point(query_vector, collection_name, top_k=2):
    # Search for the nearest point in the specified collection
    search_result = qdrant_client.search(
        collection_name=collection_name,
        query_vector=query_vector.tolist(),  # Convert NumPy array to list
        limit=top_k  # Limit the number of results
    )

    if search_result:
        for point in search_result:  # Iterate over the search result
            print(f"Closest point ID: {point.id}")  # Print point ID
            print(f"Closest point payload: {point.payload}")  # Print payload (name)
            print(f"Cosine similarity: {point.score}")  # Print distance
    else:
        print("No points found in the collection.")

# Step 3: Find the closest point to the query vector
find_closest_point(query_vector, collection_name)
```

```python
# Function to delete all collections
def delete_all_collections():
    # Get the list of all collections
    collections_response = qdrant_client.get_collections()

    # Extract collection names
    collection_names = [collection.name for collection in collections_response.collections]

    # Iterate over the collections and delete each one
    for collection_name in collection_names:
        qdrant_client.delete_collection(collection_name)  # Delete the collection
        print(f"Collection '{collection_name}' deleted.")

# Step: Delete all collections
delete_all_collections()
```
